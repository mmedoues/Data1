```python
# transform the invoice_date column to datetime type
df2['invoice_date'] = pd.to_datetime(df2['invoice_date'])
df2.dtypes
```

```python
# explaining each column meaning
# counter_statue : the status of the counter
# what is counter : a device that measures and records the quantity of a commodity or thing, especially something that cannot be directly measured.
# sort the datatypes in the counter_statue column
df2['counter_statue'].value_counts()
# drop the rows with the counter_statue column having the value of 'A'
df2 = df2[df2['counter_statue'] != 'A']
df2['counter_statue'].value_counts()
# we will change the datatype of the counter_statue column to int
df2['counter_statue'] = df2['counter_statue'].astype(int)
# now we will change the values in the dataframe and make it contain only 0 and 1 : we will keep the zeros and change all the other values to 1
df2['counter_statue'] = df2['counter_statue'].apply(lambda x: 0 if x == 0 else 1)
df2['counter_statue'].value_counts()


```

```python
# use one hot encoding to transform the counter_type column
df2 = pd.get_dummies(df2, columns=['counter_type'])
df2.head()
```

```python
# now we willchange the name of the columns coutner_type_ELEC and counter_type_GAZ to ELEC and GAZ
df2.rename(columns={'counter_type_ELEC':'ELEC', 'counter_type_GAZ':'GAZ'}, inplace=True)
df2.head()
```

```python
# now we will change the boolean values in the ELEC and GAZ columns to int
df2['ELEC'] = df2['ELEC'].astype(int)
df2['GAZ'] = df2['GAZ'].astype(int)
df2.head()
```
